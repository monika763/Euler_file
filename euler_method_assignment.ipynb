```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Function for the differential equation
def f(t, y):
    return math.cos(t) - y

# Euler method implementation
def euler_method(f, y_start, t_start, t_end, num_steps):
    h = (t_end - t_start) / num_steps
    t_points = [t_start]
    y_points = [y_start]
    t = t_start
    y = y_start
    for i in range(num_steps):
        y = y + h * f(t, y)
        t = t + h
        t_points.append(t)
        y_points.append(y)
    return t_points, y_points

# Initial values and step sizes
y0 = 1
t0 = 0
tf = 5
step_list = [20, 100, 1000]

# Different colors for each step size
colors = ['orange', 'purple', 'green']

for steps, color in zip(step_list, colors):
    t_euler, y_euler = euler_method(f, y0, t0, tf, steps)
    plt.plot(t_euler, y_euler, label=f'Steps = {steps}', color=color)

# Analytical solution (color: black)
t_fine = np.linspace(t0, tf, 1001)
y_exact = 0.5 * (np.sin(t_fine) + np.cos(t_fine) + np.exp(-t_fine))
plt.plot(t_fine, y_exact, label='Analytical Solution', color='black')

plt.xlabel('t')
plt.ylabel('y')
plt.title('Euler Method vs Analytical Solution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
import numpy as np
import pandas as pd

def f(t, y):
    return math.cos(t) - y

def euler_method(f, y_start, t_start, t_end, num_steps):
    h = (t_end - t_start) / num_steps
    t_points = [t_start]
    y_points = [y_start]
    t = t_start
    y = y_start
    for i in range(num_steps):
        y = y + h * f(t, y)
        t = t + h
        t_points.append(t)
        y_points.append(y)
    return t_points, y_points

# Parameters
initial_y = 1
start_t = 0
end_t = 5
step_sizes = [20, 100, 1000]

for steps in step_sizes:
    t_vals, y_vals = euler_method(f, initial_y, start_t, end_t, steps)
    df = pd.DataFrame({'t': t_vals, 'y_euler': y_vals})
    filename = f'euler_solution_n{steps}.csv'
    df.to_csv(filename, index=False)
```
